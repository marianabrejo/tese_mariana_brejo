```python
%store -r collection_test_set
%store -r collection_train_set
%store -r collection_pred_hw
%store -r collection_pred_sarima
%store -r collection_pred_mlp
%store -r collection_pred_lstm
```

```python
colors = ['#4477AA', '#EE6677', '#228833', '#CCBB44', '#66CCEE', '#AA3377', '#BBBBBB']
cm = 1/2.54 
```

```python
collection_train_set.plot(legend=True, label='Train', color=colors[0])
collection_test_set.plot(legend=True, label='Actual', color=colors[1])
collection_pred_hw.plot(legend=True, label='Holt-Winters', color=colors[2])
collection_pred_sarima.plot(legend=True, label='SARIMA', color=colors[3])
collection_pred_mlp.plot(legend=True, label='MLP', color=colors[4])
collection_pred_lstm.plot(legend=True, label='LSTM', color=colors[5])
plt.figsize=(17*cm,12*cm)
plt.ylabel('Monthly Supply')
plt.xlabel('Date')
plt.title("Blood Supply Predictions")
plt.legend(loc='upper left')
plt.savefig('figures/collections_results_train.pdf')
```

```python
collection_test_set.plot(legend=True, label='Actual', color='black')
collection_pred_hw.plot(legend=True, label='Holt-Winters', color=colors[0])
collection_pred_sarima.plot(legend=True, label='SARIMA', color=colors[1])
collection_pred_mlp.plot(legend=True, label='MLP', color=colors[2])
collection_pred_lstm.plot(legend=True, label='LSTM', color=colors[3])
plt.figsize=(17*cm,12*cm)
plt.ylabel('Monthly Supply')
plt.xlabel('Month')
plt.title("Monthly Supply Predictions")
plt.legend(title='Model', loc='center left', bbox_to_anchor=(1, 0.5))
plt.savefig('figures/collection_results.pdf', bbox_inches = "tight")
```

```python
#Defining MAPE function
def MAPE(Y_actual,Y_Predicted):
    mape = np.mean(np.abs((Y_actual - Y_Predicted)/Y_actual))*100
    return mape
```

### Holt-Winters

```python
np.sqrt(mean_squared_error(collection_test_set, collection_pred_hw)), mean_absolute_error(collection_test_set, collection_pred_hw), MAPE(collection_test_set, collection_pred_hw)
```

### SARIMA

```python
np.sqrt(mean_squared_error(collection_test_set, collection_pred_sarima)), mean_absolute_error(collection_test_set, collection_pred_sarima), MAPE(collection_test_set, collection_pred_sarima)
```

### MLP

```python
np.sqrt(mean_squared_error(collection_test_set, collection_pred_mlp)), mean_absolute_error(collection_test_set, collection_pred_mlp), MAPE(collection_test_set, collection_pred_mlp)
```

### LSTM

```python
np.sqrt(mean_squared_error(collection_test_set, collection_pred_lstm)), mean_absolute_error(collection_test_set, collection_pred_lstm), MAPE(collection_test_set, collection_pred_lstm)
```
