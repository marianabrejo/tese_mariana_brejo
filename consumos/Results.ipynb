```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
%store -r demand_test_set
%store -r demand_train_set
%store -r demand_prediction_hw
%store -r demand_prediction_sarima
%store -r demand_prediction_mlp
%store -r demand_prediction_lstm
```

```python
colors = ['#4477AA', '#EE6677', '#228833', '#CCBB44', '#66CCEE', '#AA3377', '#BBBBBB']
cm = 1/2.54 
```

```python
demand_train_set.plot(legend=True, label='Train', color=colors[0])
demand_test_set.plot(legend=True, label='Actual', color=colors[1])
demand_prediction_hw.plot(legend=True, label='Holt-Winters', color=colors[2])
demand_prediction_sarima.plot(legend=True, label='SARIMA', color=colors[3])
demand_prediction_mlp.plot(legend=True, label='MLP', color=colors[4])
demand_prediction_lstm.plot(legend=True, label='LSTM', color=colors[5])
plt.figsize=(17*cm,12*cm)
plt.ylabel('Demand')
plt.xlabel('Date')
plt.title("Blood Demand Predictions")
plt.legend(loc='upper left')
plt.savefig('figures/demand_results_train.pdf')
```

```python
demand_test_set.plot(legend=True, label='Actual', color='black')
demand_prediction_hw.plot(legend=True, label='Holt-Winters', color=colors[0])
demand_prediction_sarima.plot(legend=True, label='SARIMA', color=colors[1])
demand_prediction_mlp.plot(legend=True, label='MLP', color=colors[2])
demand_prediction_lstm.plot(legend=True, label='LSTM', color=colors[3])
plt.figsize=(17*cm,12*cm)
plt.ylabel('Demand')
plt.xlabel('Month')
plt.title("Monthly Demand Predictions")
plt.legend(title='Model', loc='center left', bbox_to_anchor=(1, 0.5))
plt.savefig('figures/demand_results.pdf', bbox_inches = "tight")
```

```python
#Defining MAPE function
def MAPE(Y_actual,Y_Predicted):
    mape = np.mean(np.abs((Y_actual - Y_Predicted)/Y_actual))*100
    return mape
```

### Holt-Winters

```python
np.sqrt(mean_squared_error(demand_test_set, demand_prediction_hw)), mean_absolute_error(demand_test_set, demand_prediction_hw), MAPE(demand_test_set, demand_prediction_hw)
```

### SARIMA

```python
np.sqrt(mean_squared_error(demand_test_set, demand_prediction_sarima)), mean_absolute_error(demand_test_set, demand_prediction_sarima), MAPE(demand_test_set, demand_prediction_sarima)
```

### MLP

```python
np.sqrt(mean_squared_error(demand_test_set, demand_prediction_mlp)), mean_absolute_error(demand_test_set, demand_prediction_mlp), MAPE(demand_test_set, demand_prediction_mlp)
```

### LSTM

```python
np.sqrt(mean_squared_error(demand_test_set, demand_prediction_lstm)), mean_absolute_error(demand_test_set, demand_prediction_lstm), MAPE(demand_test_set, demand_prediction_lstm)
```
