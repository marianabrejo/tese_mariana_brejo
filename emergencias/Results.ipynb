```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
%store -r emergencies_test_set
%store -r emergencies_train_set
%store -r emergencies_prediction_hw
%store -r emergencies_prediction_sarima
%store -r emergencies_prediction_mlp
%store -r emergencies_prediction_lstm
```

```python
colors = ['#4477AA', '#EE6677', '#228833', '#CCBB44', '#66CCEE', '#AA3377', '#BBBBBB']
cm = 1/2.54 
```

```python
emergencies_train_set.plot(legend=True, label='Train', color=colors[0])
emergencies_test_set.plot(legend=True, label='Actual', color=colors[1])
emergencies_prediction_hw.plot(legend=True, label='Holt-Winters', color=colors[2])
emergencies_prediction_sarima.plot(legend=True, label='SARIMA', color=colors[3])
emergencies_prediction_mlp.plot(legend=True, label='MLP', color=colors[4])
emergencies_prediction_lstm.plot(legend=True, label='LSTM', color=colors[5])
plt.figsize=(17*cm,12*cm)
plt.ylabel('Emergencies')
plt.xlabel('Date')
plt.title("Monthly Medical Emergencies Predictions")
plt.legend(title='Model', loc='center left', bbox_to_anchor=(1, 0.5))
plt.savefig('figures/emergencies_year_results.pdf', bbox_inches = "tight")
```

```python
emergencies_test_set.plot(linewidth=1.5, legend=True, label='Actual', color='black')
emergencies_prediction_hw.plot(legend=True, label='Holt-Winters', color=colors[0])
emergencies_prediction_sarima.plot(legend=True, label='SARIMA', color=colors[1])
emergencies_prediction_mlp.plot(legend=True, label='MLP', color=colors[2])
emergencies_prediction_lstm.plot(legend=True, label='LSTM', color=colors[3])
plt.figsize=(17*cm,12*cm)
plt.ylabel('Emergencies')
plt.xlabel('Month')
plt.title("Monthly Medical Emergencies Predictions")
plt.legend(title='Model', loc='center left', bbox_to_anchor=(1, 0.5))
plt.savefig('figures/emergencies_year_results_pred.pdf', bbox_inches = "tight")
```

```python
#Defining MAPE function
def MAPE(Y_actual,Y_Predicted):
    mape = np.mean(np.abs((Y_actual - Y_Predicted)/Y_actual))*100
    return mape
```

### Holt-Winters

```python
np.sqrt(mean_squared_error(emergencies_test_set, emergencies_prediction_hw)), mean_absolute_error(emergencies_test_set, emergencies_prediction_hw), MAPE(emergencies_test_set, emergencies_prediction_hw)
```

### SARIMA

```python
np.sqrt(mean_squared_error(emergencies_test_set, emergencies_prediction_sarima)), mean_absolute_error(emergencies_test_set, emergencies_prediction_sarima), MAPE(emergencies_test_set, emergencies_prediction_sarima)
```

### MLP

```python
np.sqrt(mean_squared_error(emergencies_test_set, emergencies_prediction_mlp)), mean_absolute_error(emergencies_test_set, emergencies_prediction_mlp), MAPE(emergencies_test_set, emergencies_prediction_mlp)
```

### LSTM

```python
np.sqrt(mean_squared_error(emergencies_test_set, emergencies_prediction_lstm)), mean_absolute_error(emergencies_test_set, emergencies_prediction_lstm), MAPE(emergencies_test_set, emergencies_prediction_lstm)
```
