```python
import pandas as pd
from matplotlib import rc
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.preprocessing import MinMaxScaler
from keras.preprocessing.sequence import TimeseriesGenerator
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras import layers
from tensorflow.keras import regularizers
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from keras.callbacks import EarlyStopping
from numpy import array
import seaborn as sns
```

```python
sns.set_style("darkgrid")
cm = 1/2.54 
colors = ['#4477AA', '#EE6677', '#228833', '#CCBB44', '#66CCEE', '#AA3377', '#BBBBBB']
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": ["Helvetica"]})
```

```python
df_colheitas= pd.read_csv('../colheitas/predictors/colheitas.csv', parse_dates=['date'])
df_colheitas['nr'].fillna(value=0, inplace=True)
df_colheitas=df_colheitas.groupby('date').sum()
df_colheitas=df_colheitas.resample(rule='MS').sum()
df_colheitas['type'] = 'Supply'
```

```python
df_consumos= pd.read_csv('../consumos/predictors/consumos.csv', parse_dates=['date'])
df_consumos['nr'].fillna(value=0, inplace=True)
df_consumos=df_consumos.groupby('date').sum()
df_consumos=df_consumos.resample(rule='MS').sum()
df_consumos['type'] = 'Demand'
```

```python
df = pd.concat([df_colheitas, df_consumos])
df = df.pivot_table('nr', 'date', 'type')
```

```python
df.plot(figsize=(17*cm,12*cm), linewidth=1, alpha=0.9, color=[colors[0], colors[1]], xlabel='Date', ylabel='Montly Donations and Demand')
plt.legend(title_fontsize='small', fontsize='x-small', title='', loc="lower center", bbox_to_anchor=(0.5,1), ncol=5)
plt.tight_layout()
plt.savefig('figures/supply_demand_monthly.pdf', bbox_inches = "tight")
```

### Split Train and Test

```python
aux=df
aux.drop(aux.columns[1], axis=1, inplace=True)
dataset = aux.values
dataset = aux.astype('float32')
```

```python
dataset
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
dataset = scaler.fit_transform(dataset)
```

```python
train=dataset[:240*2] 
test=dataset[240*2:]  
actual=df[240*2:]  
```

```python
len(test)
```

```python
def split_sequence(sequence, n_steps_in, n_steps_out, slide):
    X, y = list(), list()
    for i in range(0, len(sequence), slide):
        # find the end of this pattern
        end_ix = i + n_steps_in
        out_end_ix = end_ix + n_steps_out
        # check if we are beyond the sequence
        if out_end_ix > len(sequence):
            break
        # gather input and output parts of the pattern
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix:out_end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return array(X), array(y)
```

```python
raw_seq = train
n_steps_in, n_steps_out, slide = 12*2, 12*2, 1*2
X, y = split_sequence(raw_seq, n_steps_in, n_steps_out, slide)
```

```python
X[0], y[0]
```

```python
X[1], y[1]
```

### Define and fit MLP model 

```python
callback = EarlyStopping(monitor='loss', patience=100)
model = Sequential()
model.add(LSTM(units=50, activation='relu', input_shape=(n_steps_in, 1)))
model.add(Dense(n_steps_out))
model.compile(optimizer='adam', loss='mae')
model.fit(X, y, epochs=5000, callbacks=[callback])
```

### Plot losses per epoch

```python
loss_per_epoch = model.history.history['loss']
plt.plot(range(len(loss_per_epoch)),loss_per_epoch)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.figure(figsize=(17*cm,12*cm))
plt.show()
```

### Test Predicitions

```python
def MAPE(Y_actual,Y_Predicted):
    mape = np.mean(np.abs((Y_actual - Y_Predicted)/Y_actual))*100
    return mape
```

```python
def measure_error(actual, pred, step, index):
    mae = mean_absolute_error(array(actual[index:index+step]), pred)
    rmse = np.sqrt(mean_squared_error(array(actual[index:index+step]), pred))
    mape = MAPE(array(actual[index:index+step]), pred)
    return mae, rmse, mape
```

```python
x_input = train[-n_steps_in:]
x_input = x_input.reshape((1, n_steps_in, 1))
pred = model.predict(x_input).reshape(-1, 1)
pred = scaler.inverse_transform(pred)
```

```python
mae, rmse, mape = list(), list(), list()
for i in range(0, len(test), n_steps_in):
    mae_aux, rmse_aux, mape_aux = measure_error(actual, pred, n_steps_out, i)
    mae.append(mae_aux), rmse.append(rmse_aux), mape.append(mape_aux)
    x_input = array(test[i:n_steps_in+i])
    x_input = x_input.reshape((1, n_steps_in, 1))
    pred = model.predict(x_input).reshape(-1, 1)
    pred = scaler.inverse_transform(pred) 
```

```python
pred, actual
```

```python
mae,rmse,mape=list(), list(), list()
mae.append(mean_absolute_error(array(actual), pred))
rmse.append(np.sqrt(mean_squared_error(array(actual), pred)))
mape.append(MAPE(array(actual), pred))
```

```python
mape
```

```python
fig, axes = plt.subplots(1, 3, sharex=True, figsize=(17*cm,12*cm))
sns.boxplot(ax=axes[0], y=mae, color='thistle', showfliers=False)
axes[0].set_title('MAE')
sns.boxplot(ax=axes[1], y=rmse, color='lightsteelblue', showfliers=False)
axes[1].set_title('RMSE')
sns.boxplot(ax=axes[2], y=mape, color='darkseagreen', showfliers=False)
axes[2].set_title('MAPE (\%)')

#plt.savefig('plots/LSTM_boxplot.pdf')  
```
