### Split Train and Test

```python
df_diff = df[['diff']].copy()
df_diff.rename(columns = {'diff':'nr'}, inplace = True)
df_consumos = df[['nr_consumos']].copy()
df_consumos.rename(columns = {'nr_consumos':'nr'}, inplace = True)
df_colheitas = df[['nr_colheitas']].copy()
df_colheitas.rename(columns = {'nr_colheitas':'nr'}, inplace = True)
```

```python
df = pd.concat([df_colheitas, df_consumos, df_diff])
df = df.sort_values(axis=0, by=['date'])
```

```python
df
```

```python
dataset = df.values
dataset = df.astype('float32')
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
dataset = scaler.fit_transform(dataset)
```

```python
train=dataset[:240*3] 
test=dataset[240*3:]  
actual=df[240*3:]  
```

```python
def split_sequence(sequence, n_steps_in, n_steps_out, slide):
    X, y = list(), list()
    for i in range(0, len(sequence), slide):
        # find the end of this pattern
        end_ix = i + n_steps_in
        out_end_ix = end_ix + n_steps_out
        # check if we are beyond the sequence
        if out_end_ix > len(sequence):
            break
        # gather input and output parts of the pattern
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix:out_end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return array(X), array(y)
```

```python
raw_seq = train
n_steps_in, n_steps_out, slide = 12*3, 12*3, 1*3
X, y = split_sequence(raw_seq, n_steps_in, n_steps_out, slide)
```

```python
X[0], y[0]
```

```python
X[1], y[1]
```

### Define and fit MLP model 

```python
callback = EarlyStopping(monitor='loss', patience=50)
model = Sequential()
model.add(LSTM(units=50, activation='relu', input_shape=(n_steps_in, 1)))
model.add(Dense(n_steps_out))
model.compile(optimizer='adam', loss='mse')
model.fit(X, y, epochs=5000, callbacks=[callback])
```

### Plot losses per epoch

```python
loss_per_epoch = model.history.history['loss']
plt.plot(range(len(loss_per_epoch)),loss_per_epoch)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.figure(figsize=(17*cm,12*cm))
plt.show()
```

### Test Predicitions

```python
def MAPE(Y_actual,Y_Predicted):
    mape = np.mean(np.abs((Y_actual - Y_Predicted)/Y_actual))*100
    return mape
```

```python
x_input = train[-n_steps_in:]
x_input = x_input.reshape((1, n_steps_in, 1))
pred = model.predict(x_input).reshape(-1, 1)
pred = scaler.inverse_transform(pred)
```

```python
mae,rmse,mape=list(), list(), list()
mae.append(mean_absolute_error(array(actual), pred))
rmse.append(np.sqrt(mean_squared_error(array(actual), pred)))
mape.append(MAPE(array(actual), pred))
```

```python
fig, axes = plt.subplots(1, 3, sharex=True, figsize=(17*cm,12*cm))
sns.boxplot(ax=axes[0], y=mae, color='thistle', showfliers=False)
axes[0].set_title('MAE')
sns.boxplot(ax=axes[1], y=rmse, color='lightsteelblue', showfliers=False)
axes[1].set_title('RMSE')
sns.boxplot(ax=axes[2], y=mape, color='darkseagreen', showfliers=False)
axes[2].set_title('MAPE (\%)')

#plt.savefig('plots/LSTM_boxplot.pdf')  
```
